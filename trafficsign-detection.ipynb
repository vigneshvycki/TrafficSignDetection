```python
from ultralytics import YOLO
import cv2
import os
```

```python
model=YOLO('yolov8s.pt')
data_dir='/kaggle/input/cardetection/car'
train_dir=os.path.join(data_dir,'train')
val_dir=os.path.join(data_dir,'valid')
test_dir=os.path.join(data_dir,'test')
class_names=[
    'Green Light', 'Red Light', 'Speed Limit 10', 'Speed Limit 100', 'Speed Limit 110',
    'Speed Limit 120', 'Speed Limit 20', 'Speed Limit 30', 'Speed Limit 40', 'Speed Limit 50',
    'Speed Limit 60', 'Speed Limit 70', 'Speed Limit 80', 'Speed Limit 90', 'Stop'
]
dataset_yaml = f'''
path: {data_dir}
train: {train_dir}
val: {val_dir}
test: {test_dir}
names:
{os.linesep.join([f'  {i}: {name}' for i, name in enumerate(class_names)])}'''

with open('dataset.yaml','w') as f:
    f.write(dataset_yaml)
```

```python
model.train(
    data='dataset.yaml',
    epochs=30,
    imgsz=640,
    batch=8
)
```

```python
best_model_path='best_model.pt'
model.save(best_model_path)
```

```python
model=YOLO(best_model_path)
vedio_path='/kaggle/input/cardetection/video.mp4'
cap=cv2.VedioCapture(vedio_path)
output_path='output_video.mp4'
fourcc=cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))
while cap.isOpened():
    ret,frame=cap.read()
    if not ret:
        break
    result=model(frame)
    annotated_frame=result[0].plot()
    out.write(annotated_frame)
    
cap.release()
out.release()
print(f"Output video saved to {output_path}")
```
